## Birds dying :(

````python
birds, years = 80000, 0 # tuple unpacking

limit = birds/10

while birds >= limit:
    birds /= 2
    years += 1
    print(f"Year {years}: number of birds {birds:.0f} left")
````

## Got milk?

````python
import matplotlib.pyplot as plt 

bacteria = 1.5e6 # 1.5 * 10^6
factor = 1.5
hours = 0
sour = 1e7 # 10^7

bacteria_list = [bacteria]

print(bacteria_list)

while bacteria < sour:
    bacteria *= factor
    hours += 1
    bacteria_list.append(bacteria) # adds element into the list

print(bacteria_list)
print(f"Takes {hours}h for milk to get soured")

# creates a range object with numbers: 0,1,2,...,hours
hours_list = range(0, hours+1)

plt.plot(hours_list, bacteria_list, 'o-')

plt.plot(hours_list, [sour]*(hours+1)) 

plt.grid()
plt.xlabel("Hours")
plt.ylabel("Number of bacteria")
plt.title("Bacteria in milk in room temperature")
````
