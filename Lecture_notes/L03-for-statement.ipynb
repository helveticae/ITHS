### Loop through a list

```python
cute_animals = ["rabbit", "guinea pig", "cat", "guppy"]

for animal in cute_animals:
    print(f"{animal} is cute")
```

### Random module

```python
import random as rnd 

number_sixes = 0
number_rolls = 1000000

for i in range(number_rolls):
    dice = rnd.randint(1,6)
    if dice == 6:
        number_sixes += 1

print(f"{number_sixes=}")
print(f"Proportion of sixes {number_sixes/number_rolls}")

# why does it seem to converge towards 0.166...?
```

t = tiden i 친r efter 1960

a) Anta att pastakonsumtionen forts칛tter att 칬ka enligt modellen. Best칛m vilket 친rtal som den 친rliga pastakonsumtionen blir 15 kg per person.

b) Modellen st칛mmer v칛l 칬verens med verkligheten fr친n 1960 fram till idag. Utv칛rdera hur v칛l modellen kommer att st칛mma 칬verens med verkligheten i slutet av detta 친rhundrade.

```python
import math

tid, pasta = 0, 0

while pasta < 15:
    pasta = 0.791*math.e**(0.0526*tid)
    print(tid+1960, pasta)
    tid += 1
```

```python
# math.exp(1) = e

for year in range (1024):
    pasta_cons = .791*math.exp(.0526*year)
    if pasta_cons >= 15:
        print(f"칀r {year+1960} 칛r pastakonsumtionen 15 kg/person per 친r")
        break
```

```python
import math
import matplotlib.pyplot as plt

# empty list
pasta_per_year = []

# loops 100 years
for year in range(100): # year b칬rjar p친 1 eftersom range(100)
    pasta_consumtion = .791*math.exp(.0526*year)
    pasta_per_year.append(pasta_consumtion) # L칛gger till i listan f칬r varje iteration

plt.plot(range(1960, 2060), pasta_per_year)
plt.axhline(y = 15, linestyle = (0, (.5,1)))

plt.xlabel("칀r")
plt.ylabel("Pastakonsumtion i kg")
plt.title("Pastakonsumtion/kg per person per 친r fr친n 1960")
```
